```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    "Name": ["Ali", "Sara", "Ahmed", "Zara", "Bilal", "Hina", "Omar", "Ayesha"],
    "Math": [85, 90, 78, 65, 76, 59, 88, 92],
    "Physics": [80, 95, 75, 70, 78, 62, 85, 89],
    "Chemistry": [82, 93, 77, 67, 74, 60, 83, 90]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.to_csv("student_marks.csv", index=False)
```

```python
# Adding missing values
df.loc[2, "Math"] = np.nan
df.loc[[3,5], "Chemistry"] = np.nan
df
```

```python
# Filling missing values with subject mean
df["Math"].fillna(df["Math"].mean())
df["Chemistry"].fillna(df["Chemistry"].mean())
df
```

```python
# Total and Average Per Student
df["Total"] = df[["Math", "Physics", "Chemistry"]].sum(axis = 1)
df["Average"] = df[["Math", "Physics", "Chemistry"]].mean(axis = 1)
```

```python
df
```

```python
# Highest marks in each subject
print("Highest Marks:")
print(df[["Math","Physics","Chemistry"]].max())
```

```python
# Average in Each Subject
print("Class Average(Math):", df["Math"].mean())
print("Class Average(Physics):", df["Physics"].mean())
print("Class Average(Chemistry):", np.mean(df["Chemistry"])) #same as df["Chemistry"].mean()
```

```python
# Bar chart: Student total marks
plt.figure(figsize=(10,6))
plt.bar(df["Name"],df["Total"],color = ['lightblue' , 'orange'] , label = 'Students Result')
plt.title("Total Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.ylim(0 ,300)
plt.legend(loc = 'best',fontsize = 12)
plt.grid()
plt.tight_layout()
plt.savefig('total_marks.png')
plt.show()
```

```python
# Histogram: Distribution of Math scores
plt.figure(figsize=(10,6))
plt.hist(df["Math"], bins= 5, color = "orange", edgecolor = "black")
plt.title("Math Score Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.grid()
plt.tight_layout()
plt.savefig('math_score_distribution.png')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(df["Total"], labels=df["Name"], autopct="%1.1f%%", startangle=90)
plt.title("Percentage Contribution of Each Student (Total Marks)")
plt.tight_layout()
plt.savefig('total_marks_percentage.png')
plt.show()
```
